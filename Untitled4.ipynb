```python
# Importing modules  for Cleaning
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Printing information about Data Frame
df.info()
```

```python
#Removing unwanted Columns from Data frame
df.drop(columns=["Month","Status","currency","\/",'B2B',"SKU",'ship-city'],inplace = True)
df.head(10)
```

```python
# Changing the data type of Date from String to Date Format
df['Date'] = pd.to_datetime(df['Date'])
df.head(10)
```

```python
# Updating  Gender Column as "M" for Male and "W" for Women
df["Gender"] = df["Gender"].apply(lambda x : 'M' if x == "Male" else "W")
df.head(10)
```
